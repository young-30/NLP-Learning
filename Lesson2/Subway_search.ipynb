### Input data

```python
corpus = "./data/subway_station.txt"
with open(corpus, 'r', encoding='utf-8') as f:
    FILE = f.read()
```

### Process data

```python
sta_list = FILE.replace("\ufeff","").split('\n')
neighbors = list()
for line in sta_list:
    neighbor = line.split(" ")[0]
    neighbor = neighbor.replace('——', '-')
    neighbors.append(neighbor)
```

```python
len(neighbors)
```

### Extend data set

```python
# 生成每队值的相反方向值
reverse_pair = []
for line in neighbors:
        extract = line.split('-')
        new_line = extract[1] + '-' + extract[0]
        reverse_pair.append(new_line)
        

        
```

```python
neighbors_pair = neighbors + reverse_pair
```

### Finish station connection

```python
from collections import defaultdict

def build_connection(neighbor_info):
    site_list = neighbor_info
    
    site_connection = defaultdict(list)
    
    # 生成site_connect列表，连接的多个站用','分隔
    for i,line0 in enumerate(site_list):
        for line1 in site_list:
            # 整合同一站点相邻的不同站
            if line0.split('-')[0] == line1.split("-")[0] and (line1.split("-")[1] not in line0.split('-')[1]):
                line0 += (","+line1.split("-")[1])
                site_list[i] = line0
                site_list.remove(line1)
                
        
    # 生成site_connect字典
    for line in site_list:
        match_key = line.split("-")
        site_head = match_key[0]
        site_conn = match_key[1].split(',')
        site_connection[site_head] = site_conn
        
    return site_connection
```

```python
station_conn = build_connection(neighbors_pair)
```

```python
station_conn
```

### BFS search

```python
def search(start, destination, conn=station_conn):
    pathes = [[start]]  # 路径列表，里面有多条路径
    visited = set()  # 记录已遍历过的点
    
    while pathes:
        path = pathes.pop(0) # pop之后pathes值变化
        froniter = path[-1]
        
        if froniter in visited: # 如果在已经遍历的列表里，则跳过
            continue
            
        successsors = conn[froniter]
        
        for station in successsors:
            if station in path: 
                continue  # 检查回路
            
            new_path = path + [station]
            
            pathes.append(new_path)  
#             pathes = [new_path] + pathes #dfs
            
            if station == destination:
                return new_path
                
        visited.add(froniter)
```

```python
search("奥体中心", "天安门西")
```
